## NLTK Basic

```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
words = nltk.word_tokenize(text)
words
```

```python
pos_tags = nltk.pos_tag(words)
pos_tags
```

```python
chunks = nltk.ne_chunk(pos_tags, binary = True)
for chunk in chunks:
    print(chunk)
```

```python
entities =[]
labels =[]
for chunk in chunks:
    if hasattr(chunk,'label'):
        #print(chunk)
        entities.append(' '.join(c[0] for c in chunk))
        labels.append(chunk.label())
        
entities_labels = list(set(zip(entities, labels)))
entities_df = pd.DataFrame(entities_labels)
entities_df.columns = ["Entities","Labels"]
entities_df
```

```python
chunks = nltk.ne_chunk(pos_tags, binary=False) #either NE or not NE
for chunk in chunks:
    print(chunk)
    
entities =[]
labels =[]
for chunk in chunks:
    if hasattr(chunk,'label'):
        #print(chunk)
        entities.append(' '.join(c[0] for c in chunk))
        labels.append(chunk.label())
        
entities_labels = list(set(zip(entities, labels)))
entities_df = pd.DataFrame(entities_labels)
entities_df.columns = ["Entities","Labels"]
entities_df
```

## Sentence based approach

```python
entities = []
labels = []

sentence = nltk.sent_tokenize(text)
for sent in sentence:
    for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sent)),binary=False):
        if hasattr(chunk,'label'):
            entities.append(' '.join(c[0] for c in chunk))
            labels.append(chunk.label())
            
entities_labels = list(set(zip(entities,labels)))

entities_df = pd.DataFrame(entities_labels)
entities_df.columns = ["Entities","Labels"]
entities_df
```

```python
def generate_better_data(file):
    data = load_data(file)
    new_data = []
    for item in data:
        new_data.append(item)
        new_data.append(item.lower())
        new_data.append(item.upper())
    for item in data:
        name = item
        if name[-1] != 's':
            name = name + 's'
            new_data.append(name)
        
        
    data = list(set(new_data))
    return new_data
    
```

```python
def create_training_data(file,type):
    data = generate_better_data(file)
    patterns = []
    for item in data:
        pattern = {
            "label": type,
            "pattern": item
        }
        patterns.append(pattern)
    return patterns
```

## Training

```python
from spacy.tokens import DocBin
import spacy
import json
from tqdm import tqdm
import random
```

```python
def load_data(file):
    with open (file, "r", encoding="utf-8") as f:
        data = json.load(f)
    return data
```

```python
cbc_train = load_data("data/cbc_training_data.json")
```

```python
train = cbc_train[:5]
val = cbc_train[5:]
```

```python
nlp = spacy.blank("en")
def create_training(TRAIN_DATA):
    db = DocBin()
    for text, annot in tqdm(TRAIN_DATA):
        doc = nlp.make_doc(text)
        ents=[]
        for start, end, label in annot["entities"]:
            span = doc.char_span(start, end, label=label, alignment_mode="contract")
            if span is None:
                print("Skipping entity")
            else:
                ents.append(span)
        doc.ents = ents
        db.add(doc)
    return(db)
```

```python
cbc_training = create_training(train)
```

```python
cbc_val = create_training(val)
```
